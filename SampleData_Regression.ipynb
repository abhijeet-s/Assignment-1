```python
import numpy as np
import sklearn
from sklearn import datasets
import pandas as pd
```

```python
energy=pd.read_csv('NewFile4.csv') 

```

```python
Y= energy.ix[:, ['kWh']].copy()
```

```python
X = energy.ix[: ,['Month','Day','DayOfWeek','Weekday','PeakHour','temp']].copy()
```

```python
#Split the data into training and testing
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=0)

#Fit a linear regression model
from sklearn import linear_model
```

```python
lm=linear_model.LinearRegression()
```

```python
lm.fit (X_train, y_train)
```

```python
lm.coef_
```

```python
train_pred = lm.predict(X_train)
```

```python
#R-squared score of this model
from sklearn.metrics import *
r2_score(y_train, train_pred)
```

```python
test_pred = lm.predict(X_test)
```

```python
lm.intercept_
```

```python
test_pred = lm.predict(X_test)
```

```python
#Mean absolute error
mean_absolute_error(y_test, test_pred)
```

```python
#Mean squared error
mean_squared_error(y_test, test_pred)
```

```python
#Median absolute error
median_absolute_error(y_test, test_pred)
```

```python
#3.a
forecast=pd.read_csv('forecastData.csv') 
```

```python
forecast['Day'] = pd.to_datetime(forecast.ix[:, 'Day'])
```

```python
forecast = forecast.rename(columns = {'Day':'Date'})
```

```python
forecast
```

```python
forecast['Month'] = forecast['Date'].dt.month
forecast['Day'] = forecast['Date'].dt.day
forecast['Year'] = forecast['Date'].dt.year
forecast['DayOfWeek'] = forecast['Date'].dt.weekday+1
```

```python
def f(x):
    if x['DayOfWeek'] > 5:
        return 0
    else:
        return 1
```

```python
forecast['Weekday']=forecast.apply(f, axis=1)
```

```python
def f1(x):
    if x['Hr'] > 6 and x['Hr'] < 20 :
        return 1
    else:
        return 0
```

```python
forecast['PeakHour']=forecast.apply(f1, axis=1)
```

```python
cols = list(forecast)
```

```python
cols.insert(0, cols.pop(cols.index('Date')))
cols.insert(1, cols.pop(cols.index('Month')))
cols.insert(2, cols.pop(cols.index('Day')))
cols.insert(3, cols.pop(cols.index('Year')))
cols.insert(4, cols.pop(cols.index('Hr')))
cols.insert(5, cols.pop(cols.index('DayOfWeek')))
cols.insert(6, cols.pop(cols.index('Weekday')))
cols.insert(7, cols.pop(cols.index('PeakHour')))
cols.insert(8, cols.pop(cols.index('Temp')))
cols
```

```python
forecast = forecast.ix[:, cols]
forecast
```
